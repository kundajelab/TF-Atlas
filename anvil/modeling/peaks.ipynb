```python
import pandas as pd
df  = pd.read_csv('hepg2_datasets.tsv', sep='\t')
bad_1 = (df['bad_1'] == 1.0)
bad_2 = (df['bad_2'] == 1.0)
_df = df[~(bad_1 & bad_2)]
```

```python
import os
import ast
for _, row in _df.iterrows():
    experiment = row['experiment']
    peaks_url = ast.literal_eval(row['preferred_default_bed_narrowPeak_download_urls'])
    print(experiment, peaks_url)
    experiments = []
    if len(peaks_url) > 0:
        cmd = "wget {} -O /users/zahoor/lab_data3/TF-Atlas/hepG2_bed/{}.bed.gz".format(peaks_url[0], experiment)
        print(cmd)
        os.system(cmd)
        experiments.append(experiment)

#     cmd = "gsutil cp {}.bed.gz gs://tfatlas/data/{}/".format(experiment, experiment)
#     print(cmd)
#     os.system(cmd)
```

```python
import os
import ast

bed_dir = "/users/zahoor/lab_data3/TF-Atlas/hepG2_bed"
neg_bed_dir = "/users/zahoor/lab_data3/TF-Atlas/hepG2_negatives"
cnt = 0
for _, row in _df.iterrows():
    experiment = row['experiment']
    peaks_url = ast.literal_eval(row['preferred_default_bed_narrowPeak_download_urls'])
    if len(peaks_url) > 0:
        print('{}/{}.bed'.format(bed_dir, experiment))
        
        df = pd.read_csv('{}/{}.bed.gz'.format(bed_dir, experiment), sep='\t', header=None, 
                names=['chrom', 'st', 'end', 'name', 'weight', 'strand', 
                       'signal', 'p', 'q', 'summit'])
        df['weight'] = 1
        df.to_csv('{}/{}.bed'.format(bed_dir, experiment), header=None, sep='\t', index=False)

        cnt += 1

```

```python
import os
import ast

bed_dir = "/users/zahoor/lab_data3/TF-Atlas/hepG2_bed"
neg_bed_dir = "/users/zahoor/lab_data3/TF-Atlas/hepG2_negatives"
combined_dir = "/users/zahoor/lab_data3/TF-Atlas/hepG2_combined"
cnt = 0
for _, row in _df.iterrows():
    experiment = row['experiment']
    peaks_url = ast.literal_eval(row['preferred_default_bed_narrowPeak_download_urls'])
    if len(peaks_url) > 0:
        cmd = "cat {}/{}.bed {}/{}.bed > {}/{}.bed".format(bed_dir, experiment, neg_bed_dir, experiment, combined_dir, experiment)
        print(cmd)
        os.system(cmd)
        cnt += 1

```
