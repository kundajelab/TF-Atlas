```python
import pandas as pd
df  = pd.read_csv('hepg2_datasets.tsv', sep='\t')
bad_1 = (df['bad_1'] == 1.0)
bad_2 = (df['bad_2'] == 1.0)
_df = df[~(bad_1 & bad_2)]
```

```python
len(_df)
```

```python
import os
import ast
for _, row in _df.iterrows():
    experiment = row['experiment']
    peaks_url = ast.literal_eval(row['preferred_default_bed_narrowPeak_download_urls'])
    print(experiment, peaks_url)
    cmd = "wget {} -O {}.bed.gz".format(peaks_url[0], experiment)
    print(cmd)
    os.system(cmd)
    cmd = "gsutil cp {}.bed.gz gs://tfatlas/data/{}/".format(experiment, experiment)
    print(cmd)
    os.system(cmd)
```
